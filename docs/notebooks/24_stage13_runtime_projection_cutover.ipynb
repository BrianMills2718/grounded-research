# Stage 1/3 Runtime Projection Cutover

## Purpose

Define the remaining cutover from legacy `QuestionDecomposition` / `AnalystRun` runtime projections to Tyler Stage 1/3 artifacts.

## Pre-Made Decisions

1. `collect.py` consumes Tyler Stage 1 directly in the live path.
2. Fixture mode auto-detects only `tyler_stage_1.json`; legacy `decomposition.json` is explicit import-only.
3. `PipelineState.analyst_runs` is removed from the live trace.
4. A lightweight Stage 3 execution-trace surface replaces projected `AnalystRun` for debugging.
5. Semantic analyst coverage is derived from Tyler Stage 3 `AnalysisObject`s.

## Ordered Phases

### Phase 1
- Make Stage 1 the live collection contract.
- Remove `current_decomposition=` from live Stage 2 build path.

### Phase 2
- Remove live auto-loading of legacy decomposition.
- Keep legacy decomposition only behind explicit user input.

### Phase 3
- Replace `PipelineState.analyst_runs` with Stage 3 execution traces.

### Phase 4
- Make Tyler Stage 3 the only live semantic analyst contract.

### Phase 5
- Rewrite active docs and commit map.

## Acceptance

- collection consumes Tyler Stage 1 directly
- live engine no longer requires current `QuestionDecomposition`
- live engine no longer requires projected `AnalystRun`
- fixture and phase-boundary tests pass on the Tyler-native path
