# Tyler Spec Gap Audit Wave 1

## Purpose

Define a reproducible, clause-by-clause method for reviewing the live
`grounded-research` codebase against Tyler's V1 spec packet.

## Canonical Inputs

- `tyler_response_20260326/1. V1_Build_Plan_Step_By_Step (1).md`
- `tyler_response_20260326/2. V1_DESIGN (1).md`
- `tyler_response_20260326/3. V1_SCHEMAS (1).md`
- `tyler_response_20260326/4. V1_PROMPTS (1).md`
- live runtime code in `src/`, `engine.py`, `prompts/`, and `config/`

## Pre-Made Decisions

1. Tyler's packet is the authoritative contract for this audit.
2. Every finding must have a stable `spec_id` and exact code evidence.
3. Findings are classified only as literal, local divergence, extension,
   stale doc, shared-infra blocked, or Tyler ambiguity.
4. Additive local behavior may remain only if it does not conflict with
   Tyler and is documented as an extension.

## Deliverables

- `docs/TYLER_SPEC_GAP_LEDGER.md`
- clause inventory by stage
- remediation waves grouped by owner

## Acceptance

- every claimed gap maps to a Tyler clause and a live local surface
- every open gap has an owner and next action
- no prose-only "spec violation" claims survive into the final ledger
