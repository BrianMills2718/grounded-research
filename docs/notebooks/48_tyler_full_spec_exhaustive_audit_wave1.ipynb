## Execution Phases

1. Freeze the full source inventory
2. Split normative sections into audit units
3. Map audit units to local surfaces
4. Verify every unit with the minimum valid evidence type
5. Populate ledger gaps
6. Reconcile authority docs
7. Produce remediation and watch outputs


## Evidence Rules

- Static structure claims: code/config/prompt/schema inspection is acceptable
- Behavior claims: require tests, traces, CLI checks, provider verification, or runtime artifact inspection
- Findings live in the ledger; the matrix only tracks coverage


## Exit Condition

The wave closes only when every Tyler source section has been inventoried, every non-context audit unit has been audited, every real finding is ledger-backed, and the repo can answer what Tyler specified, what was checked, what still diverges, and what happens next.
