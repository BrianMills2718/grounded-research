# Wave 2 Coverage Breadth Planning

## Problem

The improved UBI bundle now completes end-to-end, but the completed report still loses the fair comparison against cached Perplexity. The report uses essentially the full available claim ledger, so the bottleneck is earlier: analyst output and raw-claim coverage are too thin for a dense, study-heavy evidence bundle.

## Current Observations

- Improved UBI bundle: 50 sources, 106 evidence items.
- Runtime-gate ledger: 20 canonical claims, 5 disputes, 2 arbitration results.
- Analyst outputs: Alpha 5 claims, Beta 6 claims, Gamma 4 claims.
- Evidence bundle contains named programs/studies including Finland, Alaska, Kenya, Stockton, Mincome, Ontario, Iran, Mongolia, Madhya Pradesh, and Namibia.
- Final claim ledger/report over-indexes Alaska and under-covers many other named cases.

## Contract

Inputs:
- Evidence bundle
- Depth profile (`analyst_claim_target`)
- Analyst reasoning frame

Outputs:
- AnalystRun with enough distinct, evidence-backed claims to reflect the breadth of a rich bundle

## Decisions

1. Use the existing `analyst_claim_target` instead of inventing a new target surface.
2. Add one corrective retry for rich bundles when returned claim count is materially below target.
3. Keep grounding strict; do not inflate breadth by allowing weaker evidence linkage.
4. Start at the analyst layer because export is already consuming the available ledger.

## Acceptance Gate

- Runtime prompt rendering includes the claim target.
- Rich-bundle under-coverage triggers one retry.
- Improved-bundle UBI rerun yields broader analyst output than the current runtime-gate run.
- Fair comparison improves, or the next narrower bottleneck becomes explicit from the completed rerun.
