# Tavily Integration Wave 1

- Status: planned
- Date: 2026-03-30
- Plan: `docs/plans/tavily_integration_wave1.md`


## Goal

Cut `grounded-research` over from the Brave-specific search tool to one shared-provider search tool backed by `open_web_retrieval`, with Tavily as the quality-first default.


## Contract

- Caller API remains `search_web(query, count, freshness, trace_id, task)`.
- Provider choice moves behind config.
- Returned JSON schema remains stable for current callers.


## Key decisions

1. Replace `tools/brave_search.py` with `tools/web_search.py`.
2. Default `collection.search_provider` to `tavily`.
3. Keep provider choice configurable without reintroducing alternate runtime families.
4. Normalize Tavily results into the existing caller-facing result schema.
