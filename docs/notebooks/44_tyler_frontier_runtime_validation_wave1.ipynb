# Tyler Frontier Runtime Validation Wave 1

Goal: run one production-config fixture-backed Tyler pipeline slice and
verify the actual model usage from run-local artifacts.


## Contracts

- Input fixture: `output/full_run/collected_bundle.json`
- Stage sidecars: `output/full_run/tyler_stage_1.json`, `output/full_run/tyler_stage_2.json`
- Output dir: `output/tyler_frontier_runtime_validation_wave1`
- Evidence of success: report, handoff, trace, observability DB
- Evidence of closure: intended stage models actually used, except valid non-dominant Stage 6 selection


## Pass / Fail

Pass:
- end-to-end run succeeds
- artifacts are written
- model usage matches the production frontier policy

Fail:
- provider/model failure blocks the run
- unexpected fallback carries a stage
- artifacts cannot prove model usage
