# Wave 1 Planning: V1 Reasoning-Quality Alignment

**Purpose:** planning artifact before implementation for the first Tyler-alignment wave.

**Status:** planned

**Related plan:** `docs/plans/v1_reasoning_quality_execution.md`


## Requirements

The first implementation wave must improve reasoning quality without changing the cheap-model development baseline.

It must:
- harden the prompt layer with Tyler's key safeguards
- introduce a dedicated claim-extraction stage
- add mechanical dedup validation and retry
- enforce anti-conformity at the schema/validator layer
- scrub analyst self-identification before downstream reuse


## Contracts To Change

### Analyst output contract
- Keep `AnalystRun` shape stable for Wave 1.
- Tighten prompt behavior, not model/provider selection.

### Claim extraction contract
- Input: `list[AnalystRun]`
- Output: `list[RawClaim]` with self-contained statements and valid evidence IDs
- New behavior: extraction becomes a dedicated post-analyst stage, not analyst copy-through

### Dedup contract
- Every raw claim must appear in exactly one group
- Zero-group outputs are invalid
- One retry allowed before fail-loud 1:1 promotion

### Arbitration update contract
- Every accepted claim update needs an allowed basis type
- Every accepted claim update needs cited evidence IDs and justification
- Invalid updates are rejected, not silently tolerated


## Execution Order

1. Prompt hardening
2. Claim extraction redesign
3. Dedup mechanical validation and retry
4. Anti-conformity schema and validator
5. Anonymization scrub
6. Tests and doc updates


## Failure Modes

- Prompt improvements preserve structure but weaken reasoning quality
- Claim extraction on cheap models becomes vaguer than current analyst claims
- Dedup retry masks systematic merge errors
- Anti-conformity validator is either toothless or over-strict
- Identity scrubbing mutates substantive content instead of just removing leaks


## Acceptance Surface

- New unit tests for claim extraction, dedup retry, anti-conformity, and anonymization
- Existing `tests/test_phase_boundaries.py` must still pass
- Prompt/contract changes must preserve the cheap-model development baseline
