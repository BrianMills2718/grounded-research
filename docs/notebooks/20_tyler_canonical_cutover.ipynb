# Tyler Canonical Cutover

- Plan: `docs/plans/tyler_canonical_cutover.md`
- Status: planned
- Goal: remove adapter debt and keep one canonical Tyler-literal runtime.


## Canonical Decisions

- One canonical contract: Tyler-literal stage artifacts.
- One canonical runtime path.
- Legacy/calibrated variants preserved by commit references and eval-time comparison, not alternate runtime modes.
- Quality-first defaults; cost-first remains configurable only as a secondary profile.


## First Deletion Slice

**Input:** question text or fixture bundle plus Tyler stage artifacts when available  
**Output:** live runtime no longer regenerates Tyler Stage 1 from `QuestionDecomposition`  
**Acceptance:** targeted tests pass and the runtime uses the Tyler Stage 1 path or regenerates directly from the question  
**Failure mode:** old fixture path depended on `decomposition.json` projection only; regenerate Stage 1 from the original question instead of restoring the adapter path.
