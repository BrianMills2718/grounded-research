# Post-Wave-2 Cleanup And Hardening

**Purpose:** planning artifact before implementing the remaining repo-local hardening work after Wave 2.

**Status:** active planning artifact

**Related plan:** `docs/plans/post_wave2_cleanup_hardening.md`


## Remaining Workstreams

1. Dense canonicalization hardening for enumeration-heavy bundles
2. Prompt/config hygiene (`collect.py`, `source_quality.py`, synthesis truncation)
3. Schema and trace construction hardening
4. Better sub-question evidence tagging


## Contract Decisions

- Enumeration-heavy canonicalization remains conservative: distinct programs/studies do not merge unless evidence context and claim content are genuinely equivalent.
- Remaining inline prompts move to YAML/Jinja2 templates rather than staying as Python strings.
- Truncation/evidence-cap policy becomes config-driven with frozen default values first.
- `AnalystRun.counterarguments` must no longer rely on fragile default/min-length behavior.
- `PhaseTrace` objects should be constructed atomically, not post-mutated.
- Sub-question evidence coverage should reflect semantic relevance, not only first-query origin.


## Verification Surface

- `PYTHONPATH=src python -m pytest tests/test_canonicalize.py`
- `PYTHONPATH=src python -m pytest tests/test_collect.py`
- `PYTHONPATH=src python -m pytest tests/test_export.py`
- `PYTHONPATH=src python -m pytest tests/test_phase_boundaries.py`

Benchmark reruns come only after the dense-canonicalization slice is mechanically verified.
